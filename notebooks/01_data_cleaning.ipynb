```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from google.colab import files

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("Libraries Loaded")
```

```python
uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print("Shape:", df.shape)

df.head()
```

```python
print("Columns:")
print(df.columns.tolist())
```

```python
df.info()
```

```python
missing = pd.DataFrame({
    "column": df.columns,
    "missing_count": df.isnull().sum().values,
    "missing_percent":
        round(
            (df.isnull().sum()/len(df))*100,
            2
        ).values
})

missing = missing.sort_values(
    "missing_percent",
    ascending=False
)

missing
```

```python
missing.to_csv(
    "missing_values_report.csv",
    index=False
)
```

```python
datetime_cols = [

    "start_datetime",
    "end_datetime",
    "closed_datetime",
    "resolved_datetime",
    "created_date"

]
```

```python
for col in datetime_cols:

    if col in df.columns:

        df[col] = pd.to_datetime(
            df[col],
            utc=True,
            errors="coerce"
        )

print("Datetime conversion complete")
```

```python
df[datetime_cols].head()
```

```python
df["duration_minutes"] = (

    df["closed_datetime"]

    -

    df["start_datetime"]

).dt.total_seconds() / 60
```

```python
negative_duration = df[
    df["duration_minutes"] < 0
]

print(
    "Negative Duration Rows:",
    len(negative_duration)
)
```

```python
df = df[
    (
        df["duration_minutes"].isna()
    )
    |
    (
        df["duration_minutes"] >= 0
    )
]

print("Shape after cleaning:", df.shape)
```

```python
df["duration_minutes"].describe()
```

```python
plt.figure(figsize=(10,6))

sns.histplot(
    df["duration_minutes"],
    bins=100
)

plt.title("Duration Distribution")
plt.xlabel("Minutes")

plt.savefig(
    "duration_distribution.png",
    bbox_inches="tight"
)

plt.show()
```

```python
if "priority" in df.columns:

    print(
        df["priority"]
        .value_counts(dropna=False)
    )
```

```python
if "priority" in df.columns:

    mode_priority = (
        df["priority"]
        .mode()[0]
    )

    df["priority"] = (
        df["priority"]
        .fillna(mode_priority)
    )

print("Priority cleaned")
```

```python
if "zone" in df.columns:

    print(
        df["zone"]
        .isnull()
        .mean()*100
    )
```

```python
if (
    "zone" in df.columns
    and
    "police_station" in df.columns
):

    station_zone_map = (

        df.groupby("police_station")["zone"]

        .agg(
            lambda x:
            x.mode().iloc[0]
            if len(x.mode()) > 0
            else np.nan
        )

    )

    df["zone"] = (

        df["zone"]

        .fillna(
            df["police_station"]
            .map(station_zone_map)
        )
    )
```

```python
if "zone" in df.columns:

    df["zone"] = (
        df["zone"]
        .fillna("Unknown Zone")
    )
```

```python
drop_cols = [

    "cargo_material",

    "age_of_truck",

    "citizen_accident_id",

    "remarks",

    "comments"

]
```

```python
existing_cols = [

    c for c in drop_cols

    if c in df.columns

]

existing_cols
```

```python
df_model = df.drop(
    columns=existing_cols,
    errors="ignore"
)

print(df_model.shape)
```

```python
duplicates = df_model.duplicated().sum()

print(
    "Duplicate Rows:",
    duplicates
)
```

```python
df_model = df_model.drop_duplicates()

print(df_model.shape)
```

```python
df_model.isnull().sum().sort_values(
    ascending=False
).head(20)
```

```python
report = []

report.append(
    f"Rows: {len(df_model)}"
)

report.append(
    f"Columns: {len(df_model.columns)}"
)

report.append(
    f"Duplicate Rows Removed: {duplicates}"
)

report.append(
    f"Negative Duration Rows Removed: {len(negative_duration)}"
)

report.append(
    f"Missing Priority After Cleaning: {df_model['priority'].isna().sum()}"
)

if "zone" in df_model.columns:

    report.append(
        f"Unknown Zone Count: {(df_model['zone']=='Unknown Zone').sum()}"
    )

with open(
    "data_quality_report.txt",
    "w"
) as f:

    for item in report:

        f.write(item + "\n")

print("\n".join(report))
```

```python
df_model.to_csv(
    "cleaned_events.csv",
    index=False
)

print(
    "Saved cleaned_events.csv"
)
```
