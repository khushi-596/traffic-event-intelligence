```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)

print("Libraries Loaded")
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print(df.shape)

df.head()
```

```python
print("Rows:", len(df))
print("Columns:", len(df.columns))
```

```python
df.info()
```

```python
cause_counts = (
    df["event_cause"]
    .value_counts()
)

cause_counts.head(15)
```

```python
plt.figure(figsize=(14,6))

cause_counts.head(15).plot(
    kind="bar"
)

plt.title(
    "Top Event Causes"
)

plt.ylabel(
    "Event Count"
)

plt.tight_layout()

plt.savefig(
    "event_cause_distribution.png"
)

plt.show()
```

```python
corridor_counts = (
    df["corridor"]
    .value_counts()
)

corridor_counts.head(15)
```

```python
plt.figure(figsize=(14,6))

corridor_counts.head(15).plot(
    kind="bar"
)

plt.title(
    "Top Corridors"
)

plt.ylabel(
    "Event Count"
)

plt.tight_layout()

plt.savefig(
    "corridor_distribution.png"
)

plt.show()
```

```python
df["priority"].value_counts()
```

```python
plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="priority"
)

plt.title(
    "Priority Distribution"
)

plt.savefig(
    "priority_distribution.png"
)

plt.show()
```

```python
hour_counts = (
    df["hour_of_day"]
    .value_counts()
    .sort_index()
)

hour_counts
```

```python
plt.figure(figsize=(14,6))

hour_counts.plot()

plt.title(
    "Events by Hour"
)

plt.xlabel(
    "Hour"
)

plt.ylabel(
    "Event Count"
)

plt.grid(True)

plt.savefig(
    "hourly_event_distribution.png"
)

plt.show()
```

```python
risk_calendar = pd.pivot_table(

    df,

    values="event_cause",

    index="corridor",

    columns="hour_of_day",

    aggfunc="count",

    fill_value=0

)
```

```python
risk_calendar.to_csv(
    "risk_calendar.csv"
)

risk_calendar.head()
```

```python
plt.figure(figsize=(18,10))

sns.heatmap(

    risk_calendar,

    cmap="Reds"

)

plt.title(
    "Corridor Risk Calendar"
)

plt.savefig(
    "corridor_hour_heatmap.png",
    bbox_inches="tight"
)

plt.show()
```

```python
cause_hour = pd.pivot_table(

    df,

    values="corridor",

    index="event_cause",

    columns="hour_of_day",

    aggfunc="count",

    fill_value=0

)
```

```python
plt.figure(figsize=(18,10))

sns.heatmap(

    cause_hour,

    cmap="YlOrRd"

)

plt.title(
    "Event Cause vs Hour"
)

plt.savefig(
    "cause_hour_heatmap.png",
    bbox_inches="tight"
)

plt.show()
```

```python
duration_heatmap = pd.pivot_table(

    df,

    values="duration_minutes",

    index="corridor",

    columns="hour_of_day",

    aggfunc="mean"

)
```

```python
plt.figure(figsize=(18,10))

sns.heatmap(

    duration_heatmap,

    cmap="Blues"

)

plt.title(
    "Average Duration by Corridor & Hour"
)

plt.savefig(
    "corridor_duration_heatmap.png",
    bbox_inches="tight"
)

plt.show()
```

```python
top_corridors = (

    df.groupby("corridor")

    .agg(

        event_count=(
            "event_cause",
            "count"
        ),

        avg_duration=(
            "duration_minutes",
            "mean"
        )

    )

    .sort_values(
        "event_count",
        ascending=False
    )

)

top_corridors.head(10)
```

```python
top_corridors.to_csv(
    "top_corridors.csv"
)
```

```python
top_causes = (

    df.groupby("event_cause")

    .agg(

        event_count=(
            "event_type",
            "count"
        ),

        avg_duration=(
            "duration_minutes",
            "mean"
        )

    )

    .sort_values(
        "event_count",
        ascending=False
    )

)

top_causes.head(10)
```

```python
top_causes.to_csv(
    "top_causes.csv"
)
```

```python
planned_analysis = (

    df.groupby("planned_flag")

    .agg(

        count=(
            "event_cause",
            "count"
        ),

        avg_duration=(
            "duration_minutes",
            "mean"
        )

    )

)

planned_analysis
```

```python
road_closure = (

    df.groupby(
        "requires_road_closure"
    )

    .agg(

        count=(
            "event_cause",
            "count"
        ),

        avg_duration=(
            "duration_minutes",
            "mean"
        )

    )

)

road_closure
```

```python
peak_hour_analysis = (

    df.groupby(
        "is_peak_hour"
    )

    .agg(

        count=(
            "event_cause",
            "count"
        ),

        avg_duration=(
            "duration_minutes",
            "mean"
        )

    )

)

peak_hour_analysis
```

```python
high_priority_corridors = (

    df[
        df["priority"]
        .astype(str)
        .str.lower()
        == "high"
    ]

    .groupby("corridor")

    .size()

    .sort_values(
        ascending=False
    )

)

high_priority_corridors.head(10)
```

```python
summary = {

    "total_events":
        len(df),

    "unique_corridors":
        df["corridor"].nunique(),

    "unique_causes":
        df["event_cause"].nunique(),

    "avg_duration":
        round(
            df["duration_minutes"].mean(),
            2
        ),

    "median_duration":
        round(
            df["duration_minutes"].median(),
            2
        )
}

summary
```

```python
summary_df = pd.DataFrame(
    [summary]
)

summary_df.to_csv(
    "eda_summary.csv",
    index=False
)
```
