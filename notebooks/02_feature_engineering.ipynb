```python
import pandas as pd
import numpy as np

pd.set_option("display.max_columns", None)

print("Libraries Loaded")
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
filename = list(uploaded.keys())[0]

df = pd.read_csv(filename)

print(df.shape)

df.head()
```

```python
datetime_cols = [

    "start_datetime",
    "end_datetime",
    "closed_datetime",
    "resolved_datetime",
    "created_date"

]
```

```python
for col in datetime_cols:

    if col in df.columns:

        df[col] = pd.to_datetime(
            df[col],
            utc=True,
            errors="coerce"
        )

print("Datetime parsed")
```

```python
df["hour_of_day"] = (
    df["start_datetime"]
    .dt.hour
)
```

```python
df["hour_of_day"].describe()
```

```python
df["day_of_week"] = (
    df["start_datetime"]
    .dt.dayofweek
)
```

```python
df["month"] = (
    df["start_datetime"]
    .dt.month
)
```

```python
df["is_weekend"] = (
    df["day_of_week"] >= 5
).astype(int)
```

```python
df["is_weekend"].value_counts()
```

```python
def create_time_bucket(hour):

    if pd.isna(hour):
        return "unknown"

    elif 4 <= hour <= 8:
        return "early_morning"

    elif 9 <= hour <= 18:
        return "daytime"

    elif 19 <= hour <= 23:
        return "night"

    return "late_night"
```

```python
df["time_bucket"] = (
    df["hour_of_day"]
    .apply(create_time_bucket)
)
```

```python
df["time_bucket"].value_counts()
```

```python
df["log_duration"] = np.log1p(
    df["duration_minutes"]
)
```

```python
df[[
    "duration_minutes",
    "log_duration"
]].head()
```

Planned vs Unplanned Flag

```python
df["planned_flag"] = (

    df["event_type"]

    .astype(str)

    .str.lower()

    .eq("planned")

).astype(int)
```

```python
df["planned_flag"].value_counts()
```

```python
if (
    "created_date" in df.columns
    and
    "start_datetime" in df.columns
):

    df["lead_time_hours"] = (

        df["start_datetime"]

        -

        df["created_date"]

    ).dt.total_seconds() / 3600

else:

    df["lead_time_hours"] = np.nan
```

```python
df.loc[
    df["lead_time_hours"] < 0,
    "lead_time_hours"
] = np.nan
```

```python
df["lead_time_hours"] = (

    df["lead_time_hours"]

    .fillna(0)

)
```

```python
if "requires_road_closure" in df.columns:

    df["requires_road_closure"] = (

        df["requires_road_closure"]

        .astype(str)

        .str.lower()

        .map({
            "yes":1,
            "true":1,
            "1":1,
            "no":0,
            "false":0,
            "0":0
        })

    )

    df["requires_road_closure"] = (

        df["requires_road_closure"]

        .fillna(0)

        .astype(int)
    )
```

```python
df["requires_road_closure"].value_counts()
```

```python
cause_frequency = (

    df["event_cause"]

    .value_counts()

)

cause_frequency.head(10)
```

```python
df["event_cause_frequency"] = (

    df["event_cause"]

    .map(cause_frequency)

)
```

```python
corridor_frequency = (

    df["corridor"]

    .value_counts()

)
```

```python
df["corridor_frequency"] = (

    df["corridor"]

    .map(corridor_frequency)

)
```

```python
zone_frequency = (

    df["zone"]

    .value_counts()

)
```

```python
df["zone_frequency"] = (

    df["zone"]

    .map(zone_frequency)

)
```

```python
peak_hours = [

    4,5,6,7,8,

    19,20,21,22,23

]
```

```python
df["is_peak_hour"] = (

    df["hour_of_day"]

    .isin(peak_hours)

).astype(int)
```

```python
df["is_peak_hour"].value_counts()
```

```python
planned_df = df[
    df["planned_flag"] == 1
].copy()
```

```python
planned_df.shape
```

```python
unplanned_df = df[
    df["planned_flag"] == 0
].copy()
```

```python
unplanned_df.shape
```
