```python
import shutil

files_to_copy = [

    "priority_classifier.pkl",

    "duration_regressor.pkl",

    "knn_model.pkl",

    "recommendation_engine.pkl",

    "feature_encoder.pkl",

    "feature_columns.pkl",

    "model_metadata.json",

    "sample_api_request.json",

    "sample_api_response.json",

    "model_performance.csv",

    "prediction_log.csv",

    "rolling_error.csv"

]
```

```python
for file in files_to_copy:

    shutil.copy(

        file,

        "deployment_assets"

    )
```

```python
import os

os.listdir(
    "deployment_assets"
)
```

```python
with zipfile.ZipFile(

    "deployment_assets.zip",

    "w",

    zipfile.ZIP_DEFLATED

) as zipf:

    for root, dirs, files_ in os.walk(
        "deployment_assets"
    ):

        for file in files_:

            path = os.path.join(
                root,
                file
            )

            zipf.write(
                path
            )
```
