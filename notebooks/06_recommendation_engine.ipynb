```python
import pandas as pd
import numpy as np

import joblib

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.neighbors import NearestNeighbors

pd.set_option("display.max_columns", None)
```

```python
df = pd.read_csv(
    "feature_matrix.csv"
)

print(df.shape)

df.head()
```

```python
priority_model = joblib.load(
    "priority_classifier.pkl"
)

duration_model = joblib.load(
    "duration_regressor.pkl"
)
```

```python
similarity_features = [

    "event_cause",

    "event_type",

    "corridor",

    "zone",

    "hour_of_day",

    "day_of_week",

    "requires_road_closure",

    "planned_flag",

    "is_peak_hour"

]
```

```python
similarity_features = [

    col

    for col in similarity_features

    if col in df.columns

]

print(similarity_features)
```

```python
MODEL_COLUMNS = [
    "event_cause",
    "event_type",
    "corridor",
    "zone",
    "hour_of_day",
    "day_of_week",
    "requires_road_closure",
    "planned_flag",
    "is_peak_hour",
    "priority_encoded" # Add priority_encoded here as it's generated and used by duration model
]
```

```python
categorical_cols = [

    col for col in [

        "event_cause",

        "event_type",

        "corridor",

        "zone"

    ]

    if col in similarity_features

]
```

```python
numerical_cols = [

    col

    for col in similarity_features

    if col not in categorical_cols

]
```

```python
from sklearn.impute import SimpleImputer

encoder = ColumnTransformer(
    transformers=[
        (
            "cat",
            OneHotEncoder(
                handle_unknown="ignore"
            ),
            categorical_cols
        ),
        (
            "num",
            SimpleImputer(strategy="mean"), # Impute numerical columns
            numerical_cols
        )
    ]
)
```

```python
X_knn = encoder.fit_transform(
    df[similarity_features]
)
```

```python
knn = NearestNeighbors(

    n_neighbors=10,

    metric="cosine"

)
```

```python
knn.fit(X_knn)
```

```python
joblib.dump(
    knn,
    "knn_model.pkl"
)
```

```python
def get_similar_events(
    event_dict,
    n_neighbors=10
):

    candidate_df = df.copy()

    # Planned vs Unplanned filter

    if (
        "planned_flag" in candidate_df.columns
        and
        "planned_flag" in event_dict
    ):

        candidate_df = candidate_df[
            candidate_df["planned_flag"]
            ==
            event_dict["planned_flag"]
        ]

    # Event Cause filter

    if (
        "event_cause" in candidate_df.columns
        and
        "event_cause" in event_dict
    ):

        event_cause = str(
            event_dict["event_cause"]
        ).lower().strip()

        candidate_df["event_cause"] = (
            candidate_df["event_cause"]
            .astype(str)
            .str.lower()
            .str.strip()
        )

        same_cause = candidate_df[
            candidate_df["event_cause"]
            ==
            event_cause
        ]

        if len(same_cause) >= 10:

            candidate_df = same_cause

    # Fallback

    if len(candidate_df) < n_neighbors:

        candidate_df = df.copy()

    candidate_encoded = encoder.transform(
        candidate_df[similarity_features]
    )

    event_df = pd.DataFrame(
        [event_dict]
    )

    event_encoded = encoder.transform(
        event_df[similarity_features]
    )

    temp_knn = NearestNeighbors(
        n_neighbors=min(
            n_neighbors,
            len(candidate_df)
        ),
        metric="cosine"
    )

    temp_knn.fit(candidate_encoded)

    distances, indices = temp_knn.kneighbors(
        event_encoded
    )

    neighbors = candidate_df.iloc[
        indices[0]
    ].copy()

    # Get a boolean mask for non-NaN duration_minutes BEFORE filtering neighbors
    notna_mask = neighbors["duration_minutes"].notna()

    # Filter neighbors using the mask
    neighbors = neighbors[notna_mask]

    # Filter distances using the same mask to maintain correspondence
    filtered_distances = distances[0][notna_mask]

    neighbors["distance"] = filtered_distances

    return neighbors
```

```python
def get_station(
    neighbors
):

    if "police_station" not in neighbors.columns:

        return "Unknown"

    station = (

        neighbors[
            "police_station"
        ]

        .dropna()

    )

    if len(station) == 0:

        return "Unknown"

    return station.mode().iloc[0]
```

```python
def manpower_band(
    neighbors
):

    avg_duration = (

        neighbors[
            "duration_minutes"
        ]
        .dropna()
        .mean()

    )

    if pd.isna(avg_duration):
        return "Medium"

    if avg_duration < 60:

        return "Low"

    elif avg_duration < 180:

        return "Medium"

    return "High"
```

```python
def diversion_route(
    neighbors
):

    candidate_cols = [

        "route_path",

        "resolved_at_address",

        "direction"

    ]

    for col in candidate_cols:

        if col in neighbors.columns:

            values = (

                neighbors[col]

                .dropna()

            )

            if len(values):

                return (
                    values.mode()
                    .iloc[0]
                )

    corridor = neighbors[
        "corridor"
    ].mode()

    if len(corridor):

        return (
            "Use alternate route near "
            + corridor.iloc[0]
        )

    return "Traffic Diversion Required"
```

```python
def similar_events_summary(
    neighbors
):

    cols = [

        c

        for c in [

            "event_cause",

            "corridor",

            "priority",

            "duration_minutes",

            "police_station"

        ]

        if c in neighbors.columns

    ]

    return (

        neighbors[cols]

        .head(5)

        .to_dict(
            orient="records"
        )

    )
```

```python
def prepare_event(event_dict, model_columns):
    event_df = pd.DataFrame([event_dict])

    # Define sensible default values for potentially missing columns
    defaults = {
        "priority_encoded": 0,  # Will be overwritten by prediction if applicable
        "month": 1,
        "event_cause_frequency": 0,
        "corridor_frequency": 0,
        "zone_frequency": 0,
        "lead_time_hours": 0,
        # These are potential columns for the duration model if trained with them
        "cause_median_duration": df["duration_minutes"].median() if "duration_minutes" in df.columns else 0,
        "corridor_median_duration": df["duration_minutes"].median() if "duration_minutes" in df.columns else 0,
    }

    # Add any missing columns that are expected by the model
    for col in model_columns:
        if col not in event_df.columns:
            if col in defaults:
                event_df[col] = defaults[col]
            else:
                # For other missing columns, fill with NaN.
                # The ColumnTransformer (encoder) and model pipelines should handle these.
                event_df[col] = np.nan

    # Ensure the order of columns matches the expected model input
    return event_df[model_columns]
```

```python
def recommend(
    event
):
    # Prepare the event DataFrame to match the training data structure
    # This function ensures all expected columns are present with appropriate defaults or NaNs.
    event_df = prepare_event(event, MODEL_COLUMNS)

    # Predict priority first using the prepared event_df
    priority_prediction = (
        priority_model
        .predict(event_df)[0]
    )

    # Add the predicted priority to event_df; this is crucial as the duration model expects it
    event_df["priority_encoded"] = priority_prediction

    # DEBUGGING PRINTS:
    print("EVENT DF COLUMNS BEFORE DURATION PREDICTION:")
    print(event_df.columns.tolist())
    print("EVENT DF HEAD BEFORE DURATION PREDICTION:")
    print(event_df.head())

    # Now predict duration using the event_df, which now includes priority_encoded
    duration_prediction = (
        duration_model
        .predict(event_df)[0]
    )

    # Transform the log-predicted duration back to original scale
    duration_prediction = np.expm1(
        duration_prediction
    )

    # Get similar events for context and additional recommendations
    neighbors = get_similar_events(
        event
    )

    # Compile all predictions and recommendations into a result dictionary
    result = {
        "predicted_priority":
            "High"
            if priority_prediction == 1
            else "Low",
        "predicted_duration_minutes":
            round(
                float(duration_prediction),
                2
            ),
        "recommended_station":
            get_station(
                neighbors
            ),
        "manpower_band":
            manpower_band(
                neighbors
            ),
        "suggested_diversion":
            diversion_route(
                neighbors
            ),
        "similar_past_events":
            similar_events_summary(
                neighbors
            )
    }

    return result
```
